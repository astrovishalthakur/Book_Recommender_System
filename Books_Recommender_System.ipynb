# Approach

## Our Stategy to Collaborative Filtering

A user "A" and user "B" reads same book "book 1".<br>
User "A" also reads another book "book 2".<br>
Now our goal is to recommend book "book 2" to user "B".

We will create a matrix.
<br>
Rows will have Users.<br>
Columns will have Books.<br>
And we will save ratings.

# Filtering Data

Now since there are more than 2 lakh books and 2 lakh users. We can't train our model on every book and every user. Since some users who are not frequent readers don't give reliable ratings.

<br>

Similarly some books are either very new or unpopular. So we can't accept rating given to every book.

<br>

Although it can be achieved using trial and error.

We will consider only those users who have given minimum 200 ratings.

Similarly we will take only those books who have received minimum of 50 ratings.

```python
book_pivot = final_rating.pivot_table(columns="user_id", index="title", values="rating")
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
 book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1, -1), n_neighbors=7)
```

```python
def get_recommendations(name, no_of_recommendations):
    num = np.where(book_pivot.index == name)[0]
    if len(num) == 0:
        print("No such book in list")
        return
    else:
        num = num[0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[num,:].values.reshape(1, -1), n_neighbors=no_of_recommendations + 1)
    return list(book_pivot.iloc[suggestions[0][1:]].index)
```

```python
recommended_list = get_recommendations("Animal Farm", 6)
```
